```python
import numpy as np
import matplotlib.pyplot as plt

def unit_step(v):
    
    if v >= 0:
        return 1
    else:
        return 0

def perceptron(x, w, b):
    
    v = np.dot(w, x) + b
    y = unit_step(v)
    return y

def AND_percep(x):
    w = np.array([1, 1])
    b = -1.5
    return perceptron(np.array(x), w, b)

# Test
example1 = [1, 1]
example2 = [1, 0]
example3 = [0, 1]
example4 = [0, 0]

x = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
y = np.array([AND_percep(example1), AND_percep(example2), AND_percep(example3), AND_percep(example4)])

print("AND({}, {}) = {}".format(example1[0], example1[1], AND_percep(example1)))
print("AND({}, {}) = {}".format(example2[0], example2[1], AND_percep(example2)))
print("AND({}, {}) = {}".format(example3[0], example3[1], AND_percep(example3)))
print("AND({}, {}) = {}".format(example4[0], example4[1], AND_percep(example4)))

# Datos de entrenamiento
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Crea la recta del perceptrón
w = np.array([1, 1])
b = -1.5
x_line = np.linspace(start=-1, stop=2)
y_line = -(w[0] * x_line + b) / w[1]

# Gráfico
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.plot(x_line, y_line, '-r')
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptrón AND')
plt.show()
```

```python

def perceptron(x, w, b):
    y = np.dot(x, w) + b
    return 1 if y >= 0 else 0

def OR_percep(x):
    w = np.array([1, 1])
    b = -0.5
    return perceptron(np.array(x), w, b)

def plot_perceptron(w, b, x_min, x_max):
    x = np.linspace(x_min, x_max, 2)
    y = -(w[0] * x + b) / w[1]
    plt.plot(x, y, label="y = -(w[0] * x + b) / w[1]")

# Test
example1 = [1, 1]
example2 = [1, 0]
example3 = [0, 1]
example4 = [0, 0]

print("OR({}, {}) = {}".format(example1[0], example1[1], OR_percep(example1)))
print("OR({}, {}) = {}".format(example2[0], example2[1], OR_percep(example2)))
print("OR({}, {}) = {}".format(example3[0], example3[1], OR_percep(example3)))
print("OR({}, {}) = {}".format(example4[0], example4[1], OR_percep(example4)))

w = np.array([1, 1])
b = -0.5
plot_perceptron(w, b, x_min=-0.5, x_max=2)
plt.scatter(1, 1, c='red', marker='x')
plt.scatter(1, 0, c='blue', marker='o')
plt.scatter(0, 1, c='blue', marker='o')
plt.scatter(0, 0, c='blue', marker='o')
plt.legend()
plt.xlim(-0.5, 2)
plt.ylim(-0.5, 2)
plt.show()
```
